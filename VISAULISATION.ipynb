```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('Impact_of_Remote_Work_on_Mental_Health.csv')
```

```python
df.head()
```

```python
# what is the distribution of work locations (remote, hybird, onsite) across different industries?
plt.style.use('ggplot')
sns.countplot(x = df['Work_Location'],palette='deep')
```

```python
plt.style.use('ggplot')
ax = sns.countplot(x = df['Work_Location'],palette='deep')
for p in ax.patches:
    ax.annotate(format(p.get_height(),'.0f'),(p.get_x() + p.get_width()/2.,p.get_height()),ha='center', va='center',xytext =(0,10), textcoords= 'offset points')
plt.show()
```

```python
fig = px.histogram(x = df['Work_Location'], text_auto=True, color=df['Work_Location'])
fig.show()
```

```python
# how do stress levels vary by work location (remote,hybird, onsite)
df['Stress_Level'].unique()
```

```python
plt.style.use('ggplot')
ax = sns.countplot(x = df['Stress_Level'],palette='deep', hue = df['Work_Location'])
for p in ax.patches:
    ax.annotate(format(p.get_height(),'.0f'),(p.get_x() + p.get_width()/2.,p.get_height()),ha='center', va='center',xytext =(0,10), textcoords= 'offset points')
plt.show()
```

```python
fig = px.histogram(x = df['Stress_Level'], text_auto=True, color=df['Work_Location'],barmode= 'group')
fig.show()
```

```python
# what is the age distribution of employees across different work locations
sns.violinplot(x=df['Work_Location'], y = df['Age'])
plt.show()
```

```python
fig = px.violin(x = df['Work_Location'], y = df['Age'], points = 'all')
fig.show()
```

```python
fig = px.box(x = df['Work_Location'], y = df['Age'], points = 'all')
fig.show()
```

```python
# what is the average social isolation rating for work location
average_social = df.groupby('Work_Location')['Social_Isolation_Rating'].mean().reset_index()
average_social
```

```python
fig = px.bar(x = average_social['Work_Location'], y = average_social['Social_Isolation_Rating'], color=average_social['Work_Location'], text=average_social['Social_Isolation_Rating'])
fig.show()
```

```python
sns.barplot(x = average_social['Work_Location'], y = average_social['Social_Isolation_Rating'], palette='deep')
plt.show()
```
