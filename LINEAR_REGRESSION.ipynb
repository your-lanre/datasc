```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import mutual_info_regression
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error, r2_score
# import missingno as mns, pip install missingno
# skilearn is moduel in changer of mancgi
```

```python
df =pd.read_csv('house_data.csv')
df.head()
```

```python
# target variable is the independent variable in a dataset 
missing_column_values = df.isnull().sum()
missing_colun_per= (df.isnull().sum()/len(df)) * 100
total_missing_values = pd.concat([missing_column_values,missing_colun_per], axis = 1,keys=['Missing values','Percentage'])
total_missing_values =total_missing_values.sort_values('Percentage', ascending= False)
total_missing_values.head(20)
```

```python
df.drop(['PoolQC','MiscFeature', 'Alley', 'Fence'], axis= 1, inplace= True)
```

```python
numerical_data = df.select_dtypes(include=['int','float'])
categorical_data = df.select_dtypes(include=['object','category'])
for x in numerical_data:
    df[x].fillna(np.mean(df[x]), inplace= True)

for x in categorical_data:
    df[x].fillna(df[x].mode()[0], inplace = True)

missing_column_values = df.isnull().sum()
missing_colun_per= (df.isnull().sum()/len(df)) * 100
total_missing_values = pd.concat([missing_column_values,missing_colun_per], axis = 1,keys=['Missing values','Percentage'])
total_missing_values =total_missing_values.sort_values('Percentage', ascending= False)
total_missing_values.head(20)
```

```python
# machine learning dosenot identify datas it only understand number by using label encoder, this how laber encoder works, first by keeping the uniques values which will  relabel it alphabe
```

```python
# pre processing
encoder = LabelEncoder()
categorical_data = df.select_dtypes(include=['object','category'])
for x in categorical_data:
    df[x] = encoder.fit_transform(df[x])
df.head()
```

```python
x =df.drop(['Id','SalePrice'],axis = 1)
y = df['SalePrice']
```

```python
# slipting data x, y into four part, 2 for x and 2 for y     #STREAMLIT
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size=0.15)
```

```python
# training the model 
model1 = LinearRegression()
model1.fit(xtrain, ytrain)
```

```python
pred1 = model1.predict(xtest)
print(mean_absolute_error(ytest, pred1))
```

```python
print(mean_absolute_percentage_error(ytest,pred1))
```

```python
print(mean_squared_error(ytest,pred1))
```

```python
print(np.sqrt(mean_squared_error(ytest, pred1)))
```
