```python
def convert_duration_to_minutes(Duration):
    if isinstance('Duration',str):
        parts = Duration.split()
        total_minutes = 0
        for part in parts:
            part = part.strip()
            if part:
                 try:
                    if 'h' in part:
                        total_minutes += int(part.replace('h',"")) * 60
                    elif 'm' in part:
                        total_minutes += int(part.replace('m',""))
                 except ValueError:
                        print(f"Unexpected format:'{part}'")
                        return total_minutes 
df['Duration'] = df['Duration'].apply(convert_duration_to_minutes)
```

```python
# z score
from scipy import stats
z_score = stats.zscore(df['Rating'])
outliers = df[(z_score > 3) | (z_score < -3)]
print(outliers)
```

```python
q1 = df['Rating'].quantile(0.25)
q3 = df['Rating'].quantile(0.75)
IQR = q3 - q1
lower = q1 - 1.5 * IQR
high = q3 + 1.5 * IQR
outlier = df[(df['Rating'] > lower) | (df['Rating'] < high)]
print(outlier)
```
